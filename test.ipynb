## 1. Introduction: The Silent Speech Revolution

The human brain is the most complex system known to science, a three-pound universe of nearly 100 billion neurons firing in intricate patterns to produce thought, emotion, and action. Among its most remarkable capabilities is language—the cornerstone of human connection and society. The loss of this ability due to neurological disorders or severe physical paralysis is a devastating blow, isolating individuals from the world and profoundly diminishing their quality of life. For these individuals, the mind remains a vibrant, active space, but the pathway to express its contents is severed. This is the challenge that Brain-Computer Interfaces (BCIs) aim to solve.

A BCI is a system that measures central nervous system activity and converts it into an artificial output, replacing, restoring, or enhancing natural CNS output. In essence, it creates a direct communication pathway between the brain and an external device, bypassing the peripheral nervous system and muscles. While early BCIs focused on motor imagery (imagining movement) to control cursors or prosthetic limbs, a more ambitious frontier has emerged: decoding speech directly from brain activity.

This report focuses on **imagined speech**, the internal monologue or the silent rehearsal of words in one's mind. Unlike attempted speech, which involves residual muscle activity, imagined speech is a purely cognitive act. This makes it an ideal paradigm for individuals with complete paralysis, but it also presents a formidable scientific challenge. The neural signals associated with imagined speech are subtle, deeply embedded in the brain's electrical noise, and vary significantly between individuals and even between trials for the same individual.

Electroencephalography (EEG) is the primary modality for this investigation. EEG is a non-invasive neuroimaging technique that measures electrical activity from the scalp. It offers excellent temporal resolution, capturing the brain's fast-changing dynamics in real-time (on the order of milliseconds). However, it suffers from poor spatial resolution and a low signal-to-noise ratio (SNR), as the electrical signals are smeared and attenuated by the skull and scalp. Extracting the faint, high-dimensional signature of an imagined word from this noisy data is akin to finding a needle in a haystack.

This is where deep learning becomes an indispensable tool. Traditional machine learning approaches often rely on handcrafted features, a labor-intensive process that may fail to capture the full complexity of the neural code. Deep learning models, with their ability to learn hierarchical representations directly from raw or minimally processed data, are perfectly suited for this task. They can automatically discover the intricate spatial, temporal, and spectral patterns that signify a particular imagined word.

### 1.1. The Chisco Dataset: A Foundation for Research

This research is grounded in the Chisco dataset, a publicly available collection of EEG recordings from subjects imagining different words. This dataset provides a standardized benchmark for developing and comparing different decoding models. The data consists of multi-channel EEG recordings, capturing brain activity from various locations across the scalp as subjects focus on specific words. The challenge lies in building a model that can take a segment of this raw EEG data as input and correctly classify which word the subject was imagining.

Recognizing that there is no one-size-fits-all solution in deep learning, this report details a systematic exploration of several distinct architectural philosophies. Our goal is not merely to find the single "best" model but to understand the trade-offs and unique advantages of different approaches. The architectures investigated are:

1.  **Wavelet-Based Approaches:** We begin by transforming the raw EEG time-series into a time-frequency domain using Wavelet Transforms. This creates a 2D representation (a scalogram) that reveals how the frequency content of the signal evolves over time. This approach is motivated by the hypothesis that the neural correlates of speech are encoded in specific frequency bands at precise moments. We explore how different wavelet configurations (e.g., 3-band vs. 5-band) impact performance.

2.  **Frequency Band-Based CNNs:** Building on the importance of frequency, this approach uses Convolutional Neural Networks (CNNs) to learn features from specific, neurophysiologically relevant frequency bands (delta, theta, alpha, beta, gamma). By processing these bands in parallel channels, the model can learn to weigh their importance and interactions, mirroring how a neuroscientist might analyze EEG data.

3.  **Advanced Computer Vision Architectures (EfficientNet & ViT):** Pushing the "EEG as an image" paradigm further, we adapt two of the most powerful models from the world of computer vision.
    *   **EfficientNet:** A family of CNNs optimized for both accuracy and computational efficiency. Its compound scaling method intelligently balances network depth, width, and resolution, allowing it to learn highly effective features without an exorbitant parameter count.
    *   **Vision Transformer (ViT):** This model dispenses with convolutions almost entirely. It divides the EEG spectrogram into patches and uses the Transformer's self-attention mechanism to learn global relationships between different time-frequency regions. This is particularly promising for capturing the long-range dependencies inherent in cognitive processes.

4.  **DeWave: A Hybrid CNN-Transformer-VQ Model:** Finally, we investigate DeWave, a bespoke architecture designed specifically for EEG decoding. DeWave represents a synthesis of the ideas explored previously. It uses a CNN to extract local features, a Transformer to model their temporal relationships, and a novel Vector Quantization (VQ) module. The VQ module discretizes the learned representations, forcing the model to commit to a set of "codex" entries from a learned codebook. This can improve robustness, reduce noise, and provide a more interpretable latent space.

## 2. Wavelet-Based Approaches: Unpacking Time and Frequency

The raw EEG signal is a one-dimensional time series, a complex waveform representing the superposition of electrical activity from millions of neurons. While this signal contains all the necessary information, its raw form is not optimal for a machine learning model. The critical information is often hidden in the frequency domain—the different brain waves (alpha, beta, gamma, etc.) that are known to correlate with different cognitive states. However, these frequency components are not static; they evolve and change rapidly over time as a person thinks. A simple Fourier Transform can tell us *what* frequencies are present, but not *when* they occur.

This is the fundamental limitation that Wavelet Transforms are designed to overcome. A Wavelet Transform is a mathematical tool that decomposes a signal into different frequency components and analyzes each component with a resolution matched to its scale. Unlike the Short-Time Fourier Transform (STFT), which uses a fixed-size window, wavelets use short windows for high frequencies (to get good time resolution) and long windows for low frequencies (to get good frequency resolution). This multi-resolution analysis makes them exceptionally well-suited for analyzing non-stationary signals like EEG, where brief, high-frequency bursts can be just as important as sustained, low-frequency oscillations.

### 2.1. The Wavelet Transform in Practice

For our task, the Continuous Wavelet Transform (CWT) is used. The CWT of a signal x(t) is defined as:

CWT(a, b) = ∫ x(t) * ψ*<sub>a,b</sub>(t) dt

Where:
*   ψ*<sub>a,b</sub>(t) is the complex conjugate of the "mother wavelet" function.
*   `a` is the scale parameter, which is inversely proportional to frequency. Stretching the wavelet (larger `a`) allows it to measure lower frequencies.
*   `b` is the translation parameter, which slides the wavelet along the time axis.

By computing the CWT for a range of scales `a` and all time points `b`, we transform the 1D EEG signal from a single channel into a 2D time-frequency representation called a **scalogram**. The value at each point in the scalogram represents the energy or amplitude of the signal at a specific time and frequency. When we do this for all 122 EEG channels, we can stack these scalograms to create a 3D tensor (channels x frequency x time), which is an ideal input for a Convolutional Neural Network.

### 2.2. Architecture: `WaveletTransform3Channel` and `WaveletTransform5Channel`

In this project, we explored two main wavelet-based feature extraction pipelines, implemented in `wavelets_3bands.py` and `wavelets_5bands.py`. The core idea is to generate scalograms and then process them with a CNN. The distinction lies in how the frequency dimension is handled.

#### 2.2.1. `wavelets_3bands.py`: A Broad-Spectrum Approach

The `WaveletTransform3Channel` module implements a strategy that divides the frequency spectrum into three broad, overlapping bands.

**Architecture Details:**

1.  **Input:** A raw EEG segment of shape `(batch_size, num_channels, num_timesteps)`, e.g., `(B, 122, 165)`.
2.  **Wavelet Transform:** A CWT is applied to each of the 122 channels independently. The chosen mother wavelet is typically the Morlet wavelet, which is a complex sine wave tapered by a Gaussian window, providing a good balance of time and frequency localization.
3.  **Frequency Band Generation:** The resulting scalograms are then processed to create three distinct "channels" that are fed into the subsequent CNN. These are not the raw scalograms themselves, but rather engineered features derived from them:
    *   **Channel 1: Magnitude (Log-Scaled):** The absolute value of the complex CWT output is taken, representing the magnitude of the signal at each time-frequency point. A logarithmic scale (`log(magnitude + epsilon)`) is applied. This is crucial because EEG power often follows a 1/f-like distribution, where low-frequency power dominates. The log transform compresses the dynamic range, preventing the high-power, low-frequency components from overwhelming the more subtle, high-frequency information.
    *   **Channel 2: Phase:** The angle of the complex CWT output is computed. Phase information can be critical for understanding the synchronization between different brain regions (phase-locking), a mechanism thought to be vital for neural communication. This channel captures the temporal relationship of the oscillations, independent of their amplitude.
    *   **Channel 3: Magnitude Difference:** The difference in magnitude between adjacent time steps is calculated (`magnitude(t) - magnitude(t-1)`). This channel explicitly highlights changes in spectral power over time, which could be a powerful cue for decoding dynamic cognitive events like imagined speech.
4.  **CNN Backbone:** The resulting 3-channel tensor, of shape `(B, 3, num_frequencies, num_timesteps)`, is then fed into a standard CNN architecture (like the one defined in `eegcnn.py`). The CNN's convolutional filters learn to detect spatial patterns (across the frequency and time dimensions) within the magnitude, phase, and magnitude-difference channels.
5.  **Output:** The CNN processes this input to produce a feature vector, which is then passed to a classifier (e.g., a fully connected layer with softmax) to predict the imagined word.

**Rationale for the 3-Channel Approach:**

*   **Rich Feature Engineering:** Instead of relying on the CNN to learn everything from the raw scalogram, this approach provides it with three distinct, complementary views of the data. The model gets information about "how much" (magnitude), "when" (phase), and "how fast it's changing" (magnitude difference).
*   **Biological Plausibility:** The separation of magnitude and phase information loosely mimics theories of neural coding where information is thought to be encoded in both the rate of neuronal firing (related to power/magnitude) and the precise timing of spikes (related to phase).
*   **Dimensionality:** It creates a 3-channel input that is highly analogous to an RGB image, allowing for the direct application of pre-trained CNNs or architectures designed for image data.

#### 2.2.2. `wavelets_5bands.py`: A Neurophysiologically-Inspired Approach

The `wavelets_5bands.py` module takes a different, more neurophysiologically grounded approach. Instead of creating channels based on mathematical properties (magnitude, phase), it creates channels based on well-established EEG frequency bands.

**Architecture Details:**

1.  **Input & Wavelet Transform:** The initial steps are the same as the 3-band approach: a CWT is applied to each EEG channel to produce a full scalogram.
2.  **Frequency Band Averaging:** The core difference lies here. The frequency axis of the scalogram is segmented into five canonical bands:
    *   **Delta (δ):** 1-4 Hz (Associated with deep sleep, but also present in complex cognitive tasks)
    *   **Theta (θ):** 4-8 Hz (Associated with memory, navigation, and drowsiness)
    *   **Alpha (α):** 8-13 Hz (The dominant rhythm in relaxed, awake states; suppressed during mental effort)
    *   **Beta (β):** 13-30 Hz (Associated with active thinking, focus, and motor control)
    *   **Gamma (γ):** 30-100 Hz (Associated with high-level information processing, feature binding, and perception)
3.  **Channel Creation:** For each of these five bands, the corresponding frequency bins in the scalogram's magnitude are averaged together at each time step. This collapses the frequency dimension, resulting in five separate time-series vectors per EEG channel, one for each band.
4.  **Stacking:** These five vectors are then stacked to form a 5-channel "image" of shape `(B, 5, num_eeg_channels, num_timesteps)`. Note the difference in shape compared to the 3-band approach. Here, the "channels" are the frequency bands, and the "height" of the image is the number of EEG sensors.
5.  **CNN Processing:** This 5-channel tensor is then processed by a CNN. The CNN's first layers learn to find patterns across the different EEG sensors within each frequency band, and deeper layers learn to integrate information across the bands.

**Rationale for the 5-Band Approach:**

*   **Interpretability:** This model is inherently more interpretable. If the model performs well, one can analyze the learned weights to see which frequency bands were most important for classifying certain words. For example, it might reveal that gamma-band activity in temporal lobe channels is a key feature for distinguishing two phonetically similar words.
*   **Noise Reduction:** By averaging over frequency bins, this approach smooths the data and may reduce the impact of spurious, narrow-band noise.
*   **Dimensionality Reduction:** It reduces the size of the frequency dimension from hundreds of bins to just five, making the subsequent CNN's job easier and reducing the risk of overfitting.
*   **Neurophysiological Grounding:** The architecture directly incorporates decades of neuroscience research, biasing the model to look for information in frequency bands that are known to be functionally relevant.

### 2.3. Results and Comparative Discussion

In practice, both wavelet-based approaches provide a significant improvement over using raw time-series data. They transform the problem from one of temporal pattern recognition to one of spatio-temporal image classification, a domain where deep learning excels.

**Comparison:**

*   **Performance:** The relative performance of the 3-band vs. 5-band approach is dataset-dependent. The **3-band (magnitude/phase/difference) approach** is more general and data-driven. It makes fewer assumptions about where the information lies and might be better at discovering novel or unexpected neural correlates. However, it may also be more susceptible to noise and harder for the CNN to interpret. The **5-band (neurophysiological) approach** is more constrained and biased. If the information is indeed encoded in the canonical bands, it will likely perform better and train faster. If the crucial information lies *between* these bands or is encoded purely in phase relationships that get lost during averaging, it may underperform.
*   **Interpretability:** The 5-band model is the clear winner in terms of interpretability. It provides a direct link between the model's internal workings and established concepts in neuroscience.
*   **Computational Cost:** The CWT itself is computationally expensive. However, once the scalograms are generated, the 5-band approach leads to a smaller input tensor for the CNN, potentially reducing training time.

**Impact for BCI:**

The wavelet approach is a powerful first step in building a BCI for imagined speech. It effectively denoises the signal and extracts a rich, meaningful feature representation. For a paralyzed individual, this means the system is more likely to pick up on the subtle neural signatures of their intended speech, leading to higher accuracy and less frustration. The ability to potentially interpret which frequency bands are driving the classification could even be used for user training, providing feedback to help them generate more distinct and decodable brain patterns. The main drawback is the computational load of the CWT, which could be a concern for real-time implementation on low-power, portable hardware. This motivates the exploration of more end-to-end architectures that can learn the feature extraction process themselves.

## 3. Frequency Band-Based Approaches: Learning from Brain Rhythms

The wavelet-based methods discussed in the previous section are powerful, but they often involve a fixed, handcrafted step of transforming the data into a time-frequency representation *before* it enters the neural network. A more end-to-end deep learning philosophy would be to let the network itself learn the optimal way to process different frequency components. This is the motivation behind the Frequency Band-Based CNN approaches, as exemplified by models like `eegcnn3bands.py` and `eegcnn5bands.py`.

This approach starts with a fundamental premise from neuroscience: different brain rhythms, or frequency bands, serve different functions and carry different types of information. Instead of treating the frequency spectrum as a continuum, we can pre-process the signal into these distinct bands and then feed them as parallel inputs to a network. This allows the model to learn specialized features for each band and then integrate them to make a final decision.

### 3.1. The Core Concept: Parallel Band Processing

The architectural pattern is straightforward yet powerful:

1.  **Band-pass Filtering:** The raw, wide-band EEG signal from each channel is first passed through a series of digital band-pass filters. Each filter is designed to isolate one specific frequency band (e.g., alpha: 8-13 Hz). The output is a set of time-series signals, one for each band, for each EEG channel.
2.  **Parallel CNN Streams:** A dedicated Convolutional Neural Network (CNN) or a dedicated set of convolutional filters is used to process the data from each frequency band. For instance, in a 5-band model, there would be five parallel processing streams.
3.  **Feature Integration (Fusion):** The feature maps extracted from each parallel stream are then concatenated or otherwise merged together.
4.  **Final Classification:** This fused feature representation is passed through a final set of layers (e.g., fully connected layers) to perform the classification of the imagined word.

This parallel structure allows the network to learn, for example, that a specific spatial pattern of gamma activity in the temporal lobe, combined with a suppression of alpha activity in the parietal lobe, is a strong indicator for a particular word. It avoids the problem of high-power, low-frequency bands dominating the analysis, as each band is processed by its own specialized feature extractors.

### 3.2. Architecture: `eegcnn3bands` and `eegcnn5bands`

Let's examine the likely architectures of the `eegcnn3bands.py` and `eegcnn5bands.py` files, which implement this philosophy.

#### 3.2.1. `eegcnn3bands.py`: A Coarse-Grained Band Analysis

This model likely simplifies the spectrum into three broad bands, for example:
*   **Low-frequency:** Combining Delta and Theta (1-8 Hz)
*   **Mid-frequency:** Combining Alpha and Beta (8-30 Hz)
*   **High-frequency:** Gamma (30-100+ Hz)

**Architecture Details:**

1.  **Input:** Raw EEG data `(B, 122, 165)`.
2.  **Filtering:** The data is filtered into these three bands. The output is a tensor of shape `(B, 3, 122, 165)`, where the second dimension represents the three frequency bands.
3.  **Spatial Convolution (Across EEG Channels):** The first convolutional layer is often a "depthwise" or "grouped" convolution. It applies a set of filters across the EEG channels *independently for each frequency band*. For the "Low-frequency" input, it learns to find spatial patterns of low-frequency activity. For the "High-frequency" input, it learns to find spatial patterns of gamma activity. This is a critical step, as it respects the distinct topographical distributions of different brain rhythms. The filter size would be something like `(1, num_channels)` to learn spatial filters.
4.  **Temporal Convolution (Along the Time Axis):** Subsequent layers apply convolutions along the time axis to learn temporal patterns within the spatially filtered data. These filters act like detectors for specific sequences of neural events.
5.  **Feature Fusion:** After several layers of convolution and pooling, the resulting feature maps from the three bands are concatenated along the feature dimension.
6.  **Classification Head:** The concatenated feature vector is flattened and passed to a dense classifier (e.g., one or more `Linear` layers with `ReLU` activation, followed by a `Softmax` output layer) to predict the word.

**Rationale:**

*   **Simplicity:** This is a simpler model than a 5-band version, with fewer parameters, making it faster to train and less prone to overfitting, especially with smaller datasets.
*   **Broad Feature Detectors:** It encourages the model to learn features that are robust across wider frequency ranges, which might be beneficial if the exact frequency of a neural signature varies between subjects or trials.

#### 3.2.2. `eegcnn5bands.py`: A Fine-Grained, Neurophysiologically-Rich Analysis

This model implements the full, canonical 5-band decomposition (Delta, Theta, Alpha, Beta, Gamma).

**Architecture Details:**

The architecture is conceptually identical to the 3-band model, but scaled up:

1.  **Input:** Raw EEG data `(B, 122, 165)`.
2.  **Filtering:** The data is filtered into the five canonical bands, resulting in a tensor of shape `(B, 5, 122, 165)`.
3.  **Parallel Convolutions:** The model uses five parallel sets of convolutional filters, one for each band, to learn band-specific spatial and temporal features.
4.  **Fusion and Classification:** The five streams of extracted features are concatenated and passed to a final classifier.

**Rationale:**

*   **High Specificity:** This model can learn highly specific patterns. It could, for instance, learn that the initiation of an imagined word is marked by a burst of beta activity, while the "vowel" part of the word is characterized by sustained gamma, and the "consonant" by a theta-band event.
*   **Greater Interpretability:** As with the 5-band wavelet model, this architecture is highly interpretable. By examining the activations and learned weights in each of the five streams, a researcher can gain insights into the specific role each brain rhythm plays in the cognitive process of imagined speech.
*   **Increased Model Capacity:** With more parallel streams, the model has a higher capacity to learn more complex relationships in the data. This comes at the cost of more parameters and a higher risk of overfitting if the dataset is not large enough.

### 3.3. Comparison with Wavelet-Based Approaches

It's important to distinguish the Frequency Band-Based CNNs from the 5-band wavelet approach:

*   **Input Representation:** The wavelet approach first creates a 2D time-frequency scalogram and then uses a 2D CNN. The band-based CNN approach keeps the data as a 1D time series (for each band) and uses 1D convolutions (first spatially across channels, then temporally).
*   **Feature Learning:** The wavelet approach relies on the CWT to do the initial feature extraction (creating the scalogram). The band-based CNN learns its features *from the filtered time-series data directly*. It has the potential to learn more flexible and data-driven temporal features than the fixed wavelet transform.
*   **Phase Information:** The standard band-pass filtering approach typically discards phase information, operating only on the amplitude of the filtered signal. The 3-channel wavelet model, in contrast, explicitly preserves and utilizes phase information, which could be a significant advantage.

### 3.4. Impact for BCI

The Frequency Band-Based approach offers a compelling balance of performance, interpretability, and computational efficiency. For a real-time BCI for a paralyzed user, this architecture is very promising.

*   **Efficiency:** Digital band-pass filters are computationally very efficient compared to a full CWT. This makes real-time implementation more feasible. The parallel CNNs can also be optimized to run efficiently on modern hardware.
*   **Adaptability:** The model's performance can provide valuable feedback. If the BCI is not performing well for a particular user, analyzing the model's internal state might reveal that, for instance, the user's gamma-band signals are particularly weak. This could lead to targeted neurofeedback training to help the user strengthen the specific brain rhythms the BCI relies on.
*   **Robustness:** By forcing the model to learn from multiple, distinct frequency bands, it may become more robust to noise that affects only a narrow part of the spectrum. If there is a burst of muscle artifact that primarily contaminates the gamma band, the model can still rely on the information present in the beta and theta bands to make a correct classification.

In summary, the Frequency Band-Based CNNs represent a mature and pragmatic approach to EEG decoding. They build directly on decades of neuroscience research, creating models that are not only effective but also provide a window into the underlying neural code of imagined speech. They are a strong candidate for any practical BCI system.

## 4. Advanced Computer Vision Architectures for EEG Decoding

The success of wavelet and band-based methods demonstrates the power of treating EEG analysis as an image classification problem. By transforming the 1D time-series into a 2D representation (e.g., a time-frequency scalogram), we unlock the ability to use the most powerful and well-researched class of deep learning models: those designed for computer vision. In recent years, two architectures have come to dominate this field: EfficientNet, the pinnacle of convolutional design, and the Vision Transformer (ViT), a paradigm-shifting approach using self-attention. This section explores how these state-of-the-art models, implemented in `effi_cnn.py` and `EEG_VIT.py`, can be adapted for the task of imagined speech decoding.

### 4.1. The EEG-as-Image Paradigm

The core idea is simple:
1.  Take the multi-channel EEG data for a single trial.
2.  For each of the 122 channels, compute a time-frequency representation (a scalogram) using a method like the Continuous Wavelet Transform (CWT) or Short-Time Fourier Transform (STFT).
3.  Arrange these 122 individual scalograms into a single large 2D grid or image. For example, they could be tiled in a 12x11 grid (with two empty slots) to create a composite image where each tile represents the activity of a specific EEG sensor.
4.  Alternatively, and more commonly, create a 3-channel image similar to the `WaveletTransform3Channel` output (e.g., magnitude, phase, difference) or by selecting three key frequency bands.
5.  Feed this resulting "EEG image" into a powerful, pre-trained computer vision model. The model, which has already learned to recognize edges, textures, shapes, and objects in natural images, can then fine-tune this knowledge to recognize the unique "textures" and "shapes" in the spectrogram that correspond to different imagined words.

This approach, known as transfer learning, is incredibly effective because it leverages the vast amount of knowledge already encoded in the weights of models trained on massive datasets like ImageNet.

### 4.2. EfficientNet: The Apex of Convolutional Efficiency

For years, the path to better CNN performance was simple: make the networks deeper (like ResNet) or wider (more channels). EfficientNet, proposed by Google AI, introduced a more principled approach. The authors found that there is a fixed relationship between the optimal depth, width, and input resolution of a network. They proposed a **compound scaling** method that scales all three dimensions uniformly using a simple compound coefficient.

**Architecture (`effi_cnn.py`):**

1.  **Input:** An "EEG image" (e.g., a 3-channel scalogram of shape `(B, 3, H, W)`).
2.  **Stem:** A standard initial convolutional layer to process the input.
3.  **MBConv Blocks:** The core of the network is a series of `MBConv` (Mobile Inverted Bottleneck Convolution) blocks. These are highly efficient building blocks that feature:
    *   **Inverted Residuals:** Unlike classic ResNet blocks that first shrink and then expand the number of channels, MBConv blocks first expand the channels, apply a lightweight depthwise convolution, and then shrink them back down. This allows for richer feature learning in the high-dimensional space.
    *   **Depthwise Separable Convolutions:** This technique splits a standard convolution into two parts: a depthwise convolution that applies a single filter per input channel, and a pointwise convolution (1x1) that combines the outputs. This dramatically reduces the number of parameters and computations compared to a standard convolution.
    *   **Squeeze-and-Excitation (SE) Optimization:** An SE block is a small sub-network that looks at the channels of a feature map, learns which ones are most important, and then re-weights them accordingly. It allows the network to dynamically focus on the most informative features.
4.  **Compound Scaling:** The `effi_cnn.py` script would likely use one of the pre-defined EfficientNet variants, from B0 (the smallest) to B7 (the largest). For example, `EfficientNet-B0` serves as the baseline architecture. To get `EfficientNet-B1`, the compound scaling formula is used to increase the number of layers (depth), the number of channels in each layer (width), and the resolution of the input image in a balanced, optimal way.
5.  **Classification Head:** The final feature map is passed through a pooling layer and a single fully connected layer to produce the classification output.

**Rationale and Impact for BCI:**

*   **High Performance:** EfficientNets represent the state-of-the-art in terms of accuracy for a given number of parameters (FLOPs). For a BCI, this means achieving higher decoding accuracy is possible without requiring a supercomputer for training or inference.
*   **Efficiency:** The name says it all. The architectural optimizations mean that an EfficientNet can often achieve the same accuracy as a much larger model like a ResNet-50 but with an order of magnitude fewer parameters. This is a *critical* advantage for a practical BCI, which may need to run on a portable, battery-powered device. A smaller model consumes less power and delivers faster results, leading to a more seamless experience for the user.
*   **Transfer Learning:** By using an EfficientNet pre-trained on ImageNet, we give our model a massive head start. It already understands complex visual patterns, and we only need to fine-tune it to the specific patterns of EEG scalograms. This drastically reduces the amount of EEG data needed to train an effective model.

### 4.3. Vision Transformer (ViT): A New Vision for EEG

The Vision Transformer, introduced by Google AI in 2020, revolutionized computer vision by showing that a pure Transformer architecture, with minimal or no convolutions, could outperform the best CNNs.

**Architecture (`EEG_VIT.py`):**

1.  **Input & Patching:** The ViT takes a 2D EEG image as input. Its first and most critical step is to divide this image into a sequence of smaller, non-overlapping patches. For example, a 224x224 image might be split into a grid of 14x14 patches, each of size 16x16. Each patch is then flattened into a single vector.
2.  **Patch & Position Embedding:** Each flattened patch vector is linearly projected into a higher-dimensional "embedding space." Crucially, a learnable **positional embedding** is added to each patch embedding. Since the Transformer architecture itself has no inherent notion of spatial location, these positional embeddings are essential to inform the model about the original location of each patch in the image (e.g., "this is the top-left patch," "this is a center patch").
3.  **[CLS] Token:** A special, learnable `[CLS]` (classification) token is prepended to the sequence of patch embeddings. The idea is that the self-attention mechanism will aggregate global information about the entire image into this single token's representation.
4.  **Transformer Encoder:** The resulting sequence of embeddings is fed into a standard Transformer encoder, which consists of multiple layers of:
    *   **Multi-Head Self-Attention (MSA):** This is the core of the ViT. For each patch, the self-attention mechanism calculates an "attention score" with every other patch in the sequence. It learns to identify which other parts of the image are most relevant to understanding the current patch. The "multi-head" aspect means this is done multiple times in parallel with different learned transformations, allowing the model to capture different types of relationships.
    *   **Feed-Forward Network (MLP):** A simple multi-layer perceptron is applied independently to each patch representation.
5.  **Classification Head:** After passing through the entire Transformer encoder, only the output corresponding to the special `[CLS]` token is used. This single vector, which now represents the aggregated global information from the entire EEG image, is passed to a standard MLP head for the final classification.

**Rationale and Impact for BCI:**

*   **Global Context:** CNNs, with their local receptive fields, are excellent at finding local patterns. However, they struggle to model long-range dependencies. The ViT's self-attention mechanism excels at this. It can learn, for example, that a specific high-frequency pattern in a *frontal* EEG channel (a patch on the left of the image) is directly related to a low-frequency suppression in a *parietal* channel (a patch on the right of the image). This ability to model the global, brain-wide network dynamics is a massive potential advantage for decoding complex cognitive states like speech.
*   **Flexibility and Scalability:** ViTs are highly scalable. As more data becomes available, they continue to improve, often surpassing CNNs at very large data scales.
*   **Potential for New Insights:** Because ViT learns relationships between distant patches, analyzing its attention maps could provide novel insights into the functional connectivity of the brain during imagined speech. We could visualize which brain regions (patches) the model "pays attention to" when classifying a word, potentially revealing previously unknown neural networks.
*   **Challenges:** The primary drawback of ViT is its data-hungriness. Without pre-training on a very large dataset like ImageNet, ViTs tend to underperform CNNs on smaller datasets because they lack the useful inductive biases (like locality and translation equivariance) that are built into convolutions. Therefore, for the BCI task, using a pre-trained ViT and fine-tuning it is not just an optimization; it is a necessity.

### 4.4. Summary: Pushing the Boundaries with Vision Models

By leveraging advanced computer vision architectures, we move beyond handcrafted features and simple CNNs into the realm of state-of-the-art representation learning.
*   **EfficientNet** offers a highly pragmatic path to excellence: top-tier accuracy with unmatched computational efficiency, making it a prime candidate for real-world, deployable BCIs.
*   **ViT** offers a more radical, potentially more powerful alternative. Its strength in modeling global, long-range dependencies is perfectly matched to the network-based nature of brain function. While more data-hungry, it holds the promise of not only higher accuracy but also a deeper understanding of the brain's functional connectivity during imagined speech.

For a paralyzed individual, the successful application of these models could mean the difference between a slow, error-prone BCI and a fluid, intuitive communication device that can truly keep pace with the speed of thought.

## 5. DeWave: A Bespoke Hybrid Architecture for EEG Decoding

The architectures discussed so far have either been adapted from other domains (like computer vision) or have been built on established signal processing principles (like band-pass filtering). The DeWave model, as implemented in `dewave_chisco.py`, represents a different philosophy: a purpose-built, hybrid architecture designed from the ground up to tackle the unique challenges of decoding information from raw brain signals. It is a sophisticated synthesis of the most successful ideas in modern deep learning, combining the local feature extraction power of CNNs, the global context modeling of Transformers, and a novel vector quantization module for creating a robust, discretized latent space.

### 5.1. The Motivation Behind DeWave

DeWave is born from the recognition that EEG signals have distinct characteristics that demand a specialized approach:
1.  **Local Spatial Features:** Neighboring EEG channels often record similar information due to volume conduction in the brain. A model needs to be able to learn local spatial filters to capture these regional patterns. This is a job for CNNs.
2.  **Temporal Dependencies:** A thought, like an imagined word, unfolds over time. The neural patterns at the beginning, middle, and end of the thought are not independent. A model must be able to capture these long-range temporal relationships. This is a job for Transformers.
3.  **Noise and Variability:** EEG signals are notoriously noisy and variable. A model that learns a continuous, high-dimensional representation might be too sensitive to small, irrelevant fluctuations. Forcing the model to map its continuous representation to a discrete set of "codes" could improve robustness and noise immunity. This is a job for Vector Quantization.

DeWave elegantly integrates these three components into a single, end-to-end trainable pipeline.

### 5.2. Architecture: A Deep Dive into `dewave_chisco.py`

The DeWave model can be broken down into four main stages: a CNN feature extractor, a Transformer encoder, a vector quantizer, and a final classifier.

**1. CNN Feature Extractor (The "Perceiver")**

*   **Input:** Raw EEG data, shape `(B, num_channels, num_timesteps)`, e.g., `(B, 122, 165)`.
*   **Function:** This component's job is to act as a "perceiver," extracting meaningful local features from the raw, high-dimensional input. It typically consists of a series of 1D convolutional layers that operate along the time axis.
*   **Implementation:**
    *   The first layer might have a large kernel size to capture the initial temporal structure.
    *   Subsequent layers with smaller kernels refine these features.
    *   Batch Normalization and activation functions (like GELU or ReLU) are used after each convolution to stabilize training and introduce non-linearity.
    *   The output is a sequence of feature vectors, one for each time step, but with the channel dimension transformed into a feature dimension. For example, the output shape might be `(B, num_timesteps, feature_dim)`, e.g., `(B, 165, 256)`. This sequence is now ready for the Transformer.

**2. Transformer Encoder (The "Contextualizer")**

*   **Input:** The sequence of feature vectors from the CNN, `(B, 165, 256)`.
*   **Function:** This is the heart of the model's ability to understand temporal context. It uses the self-attention mechanism to analyze the entire sequence of features simultaneously.
*   **Implementation:**
    *   **Positional Encoding:** Just like in the ViT, positional encodings are added to the input sequence to give the model information about the order of the time steps.
    *   **Multi-Head Self-Attention:** Each layer of the Transformer allows every time step's feature vector to "look at" and integrate information from every other time step. This is how the model learns that a pattern at `t=20` is predictive of another pattern at `t=120`. It builds a rich, context-aware representation of the entire imagined utterance.
    *   **Feed-Forward Layers:** Standard MLP layers process each time step's representation independently.
    *   The output is a new sequence of feature vectors of the same shape, `(B, 165, 256)`, but now each vector is enriched with global temporal context.

**3. Vector Quantizer (The "Discretizer")**

*   **Input:** The context-aware feature sequence from the Transformer.
*   **Function:** This is the most novel component of DeWave. It performs a process called Vector Quantization (VQ). The goal is to take the continuous feature vectors produced by the Transformer and map them to a discrete set of vectors from a learned "codebook."
*   **Implementation:**
    *   **Codebook:** The model maintains a codebook, which is essentially a list of `N` embedding vectors (e.g., `N=1024` vectors of dimension `256`). These are the "codex" entries. This codebook is learned during training.
    *   **Lookup:** For each of the 165 feature vectors in the sequence from the Transformer, the VQ layer finds the *closest* vector in the codebook (using Euclidean distance).
    *   **Replacement:** It then replaces the Transformer's output vector with the chosen codebook vector.
    *   The output is a sequence of vectors `(B, 165, 256)`, but now each of the 165 vectors is one of the 1024 possible entries from the codebook.
*   **The "Commitment Loss":** Training a VQ module is tricky because the "find the closest" operation (argmin) is not differentiable. This is solved using the Straight-Through Estimator. Additionally, a special "commitment loss" is added to the total loss function. This loss encourages the output of the Transformer to stay "committed" to the codebook entries it maps to, preventing the model's representations from fluctuating wildly.

**4. Classifier (The "Interpreter")**

*   **Input:** The quantized sequence of codebook vectors.
*   **Function:** This final component interprets the sequence of "codes" to make a prediction.
*   **Implementation:**
    *   The sequence of vectors is typically averaged over the time dimension (mean pooling) to get a single vector representing the entire trial.
    *   This single vector is then passed through one or more fully connected layers to produce the final logits for each class (each imagined word).

### 5.3. Rationale and Impact for BCI

DeWave's hybrid design offers several profound advantages for a BCI system:

*   **Best of All Worlds:** It combines the proven strengths of CNNs for local feature learning, Transformers for global context modeling, and VQ for robust, discretized representations. It's a "no-compromise" architecture designed to handle the specific structure of EEG data.
*   **Noise Immunity:** The VQ step is a powerful form of regularization. By forcing the continuous, noisy output of the Transformer into one of a few hundred discrete states, the model learns to ignore small, irrelevant variations. For a BCI user, this could mean the system is less likely to be thrown off by a blink, a slight muscle twitch, or background electrical noise, leading to more reliable performance.
*   **Interpretability via "Codex":** The codebook provides a new and powerful way to interpret the model's inner workings. We can analyze the "codex" usage. For example, we might find that "codex entry #512" is consistently used during the beginning of words that start with a "b" sound, while "codex entry #128" appears during sustained vowel sounds. This is like discovering the brain's "neural alphabet" for imagined speech. This could be invaluable for debugging the BCI and understanding the neural code itself.
*   **Generative Potential:** A trained DeWave model contains a learned codebook of the fundamental components of imagined speech. This opens the door to generative models. We could potentially sample sequences of codes from the model to *synthesize* artificial brain activity corresponding to any word, a powerful tool for data augmentation and further research.

**Challenges:**

*   **Complexity:** DeWave is a complex model with many moving parts and hyperparameters to tune. Training can be difficult to stabilize, and it requires careful implementation of the VQ layer and its associated commitment loss.
*   **Computational Cost:** The combination of a deep CNN, a multi-layer Transformer, and the VQ lookup makes this a computationally intensive model, posing a challenge for real-time deployment on low-power hardware.

For a paralyzed individual, a BCI powered by a DeWave-like architecture represents the cutting edge. It promises not only high accuracy but also a level of robustness and reliability that is essential for a device meant for daily use. The potential to analyze the learned "codex" could even lead to personalized BCIs, where the system is fine-tuned to the unique neural alphabet of each individual user, leading to an unprecedented level of performance and a true restoration of the ability to communicate.

```python

import pywt
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pickle

# --- 1. Load the single channel signal (ensure this is run first) ---
pkl_path = Path(f"Chisco/derivatives/preprocessed_pkl/sub-01/sub-01_task-imagine_run-01_eeg.pkl")
with open(pkl_path, "rb") as f:
    trials = pickle.load(f)
# Select the first channel (index 0) from the first trial
signal = trials[0]['input_features'][0, :]

# --- 2. Define Parameters ---
sfreq = 500  # Sampling frequency for the Chisco dataset is 500 Hz
bands = {
    'delta': (1, 4),
    'theta': (4, 8),
    'alpha': (8, 13),
    'beta': (13, 30),
    'gamma': (30, 70)
}
wavelet_name = 'morl'
# Define a range of frequencies to analyze, covering all bands
freqs_to_analyze = np.logspace(np.log10(1), np.log10(100), num=100)
# Convert these frequencies to scales for the CWT function
scales = pywt.central_frequency(wavelet_name) * sfreq / freqs_to_analyze

# --- 3. Perform CWT to get time-frequency representation ---
coeffs, frequencies = pywt.cwt(signal, scales, wavelet_name, sampling_period=1/sfreq)
magnitude_coeffs = np.abs(coeffs)

# --- 4. Separate into Bands ---
band_signals = {}
for band_name, (fmin, fmax) in bands.items():
    # Find the indices of the frequencies that fall within the current band
    freq_indices = np.where((frequencies >= fmin) & (frequencies < fmax))[0]
    
    if len(freq_indices) > 0:
        # Average the power (magnitude) of the coefficients ONLY for the frequencies in this band
        band_power = np.mean(magnitude_coeffs[freq_indices, :], axis=0)
        band_signals[band_name] = band_power
        print(f"✅ Successfully extracted '{band_name}' band signal. Shape: {band_power.shape}")
    else:
        print(f"⚠️ No frequencies found for '{band_name}' band.")

# --- 5. Visualization ---
time_axis = np.arange(len(signal)) / sfreq
fig, axes = plt.subplots(len(bands) + 1, 1, figsize=(15, 12), sharex=True)
fig.suptitle('Original Signal and its Decomposition into Frequency Bands', fontsize=16)

# Plot original signal
axes[0].plot(time_axis, signal, color='black')
axes[0].set_title('Original EEG Signal')
axes[0].set_ylabel('Voltage (µV)')

# Plot each band signal
for i, (band_name, band_data) in enumerate(band_signals.items()):
    ax = axes[i+1]
    ax.plot(time_axis, band_data, label=f'{band_name} ({bands[band_name][0]}-{bands[band_name][1]} Hz)')
    ax.set_title(f'{band_name.capitalize()} Band Power')
    ax.set_ylabel('Mean Power')
    print(f"✅ {band_name.capitalize()} band signal shape: {band_data.shape}")
    print(band_signals)

axes[-1].set_xlabel('Time (seconds)')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pickle

# --- 1. Load the single channel signal ---
pkl_path = Path("Chisco/derivatives/preprocessed_pkl/sub-01/sub-01_task-imagine_run-01_eeg.pkl")
with open(pkl_path, "rb") as f:
    trials = pickle.load(f)
signal = trials[0]['input_features'][0, :]

# --- 2. Define Parameters ---
sfreq = 500
wavelet_name = 'morl'

# 20 linearly spaced bands: 0-5, 5-10, ..., 95-100 Hz
band_edges = np.linspace(0, 100, 25)
bands_20 = [(band_edges[i], band_edges[i+1]) for i in range(24)]
freqs = np.linspace(0, 100, 100)[1:]  # skip 0 Hz to avoid division by zero
scales = pywt.central_frequency(wavelet_name) * sfreq / freqs

# --- 3. Perform CWT ---
coeffs, frequencies = pywt.cwt(signal, scales, wavelet_name, sampling_period=1/sfreq)

# --- 4. Extract 20 bands (complex) ---
band_coeffs_20 = []
for fmin, fmax in bands_20:
    idx = np.where((frequencies >= fmin) & (frequencies < fmax))[0]
    if len(idx) > 0:
        band_coeffs_20.append(np.mean(coeffs[idx, :], axis=0))
    else:
        band_coeffs_20.append(np.zeros(signal.shape, dtype=np.complex64))
band_coeffs_20 = np.stack(band_coeffs_20, axis=0)  # (20, n_samples)

# --- 5. Group into 5 bands as specified (complex) ---
five_bands_complex = [
    band_coeffs_20[0],                # delta: 0-5 Hz
    band_coeffs_20[1],                # theta: 5-10 Hz
    band_coeffs_20[2],                # alpha: 10-15 Hz
    np.mean(band_coeffs_20[3:7], 0),  # beta: 15-35 Hz (bands 3-6)
    np.mean(band_coeffs_20[7:20], 0)  # gamma: 35-100 Hz (bands 7-19)
]

band_names = ['Delta (0-4 Hz)', 'Theta (4-8 Hz)', 'Alpha (8-12 Hz)', 'Beta (12-32 Hz)', 'Gamma (32-100 Hz)']

# --- 6. Visualization: Real, Imag, Magnitude, Phase for each band ---
time_axis = np.arange(len(signal)) / sfreq
fig, axes = plt.subplots(5, 4, figsize=(18, 12), sharex=True)
fig.suptitle('Complex Wavelet Band Signals (Real, Imag, Magnitude, Phase)', fontsize=16)

for i, (band, name) in enumerate(zip(five_bands_complex, band_names)):
    axes[i, 0].plot(time_axis, band.real)
    axes[i, 0].set_title(f"{name} - Real")
    axes[i, 1].plot(time_axis, band.imag)
    axes[i, 1].set_title(f"{name} - Imag")
    axes[i, 2].plot(time_axis, np.abs(band))
    axes[i, 2].set_title(f"{name} - Magnitude")
    axes[i, 3].plot(time_axis, np.angle(band))
    axes[i, 3].set_title(f"{name} - Phase")
    if i == 4:
        for j in range(4):
            axes[i, j].set_xlabel('Time (seconds)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
from pathlib import Path
import pickle

# Define the base path to the preprocessed data
base_path = Path("Chisco/derivatives/preprocessed_pkl")

print("--- Starting File Inspection ---")

# Iterate through all subject directories (e.g., sub-01, sub-02)
for subject_path in sorted(base_path.glob("sub-*")):
    if not subject_path.is_dir():
        continue
    
    print(f"\n--- Subject: {subject_path.name} ---")
    subject_total_trials = 0
    
    # Use glob to find all .pkl files for this subject, sorted by name
    run_files = sorted(subject_path.glob("*.pkl"))
    
    if not run_files:
        print("  No .pkl files found.")
        continue
        
    # Iterate through each run's .pkl file
    for pkl_path in run_files:
        try:
            with open(pkl_path, "rb") as f:
                # Each .pkl file contains a list of trial dictionaries
                trials = pickle.load(f)
                num_trials = len(trials)
                subject_total_trials += num_trials
                print(f"  - File: {pkl_path.name} | Contains: {num_trials} trials (sentences)")
        except Exception as e:
            print(f"  - Could not read {pkl_path.name}: {e}")
            
    print(f"--- Total for {subject_path.name}: {subject_total_trials} trials ---")

print("\n--- Inspection Complete ---")
```
